```python
import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load your dataset
dataset_path = "/kaggle/input/unplash-dataset/dataset_final.csv"
df = pd.read_csv(dataset_path)

# Define the training data
training_data = df["photo_description"].tolist()  
training_text = " ".join(training_data) 

# Initialize the GPT-2 model and tokenizer
model_name = "gpt2"  
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Tokenize the training data
input_ids = tokenizer(training_text, return_tensors="pt", padding=True, truncation=True, max_length=128)



# Generate text prompts using your trained model
prompts = ["Describe a serene landscape with a clear blue sky.",
           "Write a creative story about a bustling city at night."]
   
for prompt in prompts:
    input_ids = tokenizer(prompt, return_tensors="pt")
    output = model.generate(input_ids["input_ids"], max_length=100, num_return_sequences=1)
    generated_prompt = tokenizer.decode(output[0], skip_special_tokens=True)
    print(generated_prompt)
```

```python
import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load your dataset
dataset_path = "/kaggle/input/unplash-dataset/dataset_final.csv"
df = pd.read_csv(dataset_path)

# Convert "photo_description" column to strings
df["photo_description"] = df["photo_description"].astype(str)

# Define the training data
training_data = df["photo_description"].tolist()  
training_text = " ".join(training_data)  

# Initialize the GPT-2 model and tokenizer
model_name = "gpt2"  
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# Tokenize the training data
input_ids = tokenizer(training_text, return_tensors="pt", padding=True, truncation=True, max_length=128)



# Generate text prompts using your trained model
prompts = ["Describe a serene landscape with a clear blue sky.",
           "Write a creative story about a bustling city at night."]
   
for prompt in prompts:
    input_ids = tokenizer(prompt, return_tensors="pt")
    output = model.generate(input_ids["input_ids"], max_length=100, num_return_sequences=1)
    generated_prompt = tokenizer.decode(output[0], skip_special_tokens=True)
    print(generated_prompt)

    
```
